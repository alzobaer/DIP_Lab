Importing the libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Function to perform enhancement

```python
def enhance_image(image, max, min, add_intensity):
    enhanced_image = image.copy()
    height, width = enhanced_image.shape

    for row in range(height):
        for col in range(width):
            if enhanced_image[row][col] >= min and enhanced_image[row][col] <= max:
                enhanced_image[row][col] += add_intensity
                if enhanced_image[row][col] > 255:
                    enhanced_image[row][col] = 255
    
    return enhanced_image
            
```

Load the original grayscale image

```python
original_image = cv2.imread('./images/moon.jpg', 0)
grayscale_image = cv2.resize(original_image, (512, 512))
```

Define the range of gray levels to enhance (e.g., 50 to 150)

```python
max = 50
min = 0
add_intensity = 80
```

Perform enhancement

```python
enhanced_image = enhance_image(grayscale_image, max, min, add_intensity)
```

Displaying the original image and enhanced image

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(grayscale_image, cmap='gray')
plt.title('The original image')

plt.subplot(1, 2, 2)
plt.imshow(enhanced_image, cmap='gray')
plt.title('The original image')

plt.tight_layout()
plt.show()
```
