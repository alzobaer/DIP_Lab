Importing the libraries

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

Function for Bit Plane Slicing

```python
def make_sclice(image):
    mask = 11100000     # Keep only the last three bits (most significant bits)
    last_three_bits_image = image & mask

    return last_three_bits_image
```

Loading the original image

```python
original_image = cv2.imread('./images/img1.webp', 0)
gray_image = cv2.resize(original_image, (512, 512))
```

Displaying the original image, last 3 bits image and difference image

```python
last_three_bits_image = make_sclice(gray_image)
difference_image = gray_image - last_three_bits_image

plt.figure(figsize=(12, 12))
plt.subplot(2, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('The Original Image')

plt.subplot(2, 2, 2)
plt.imshow(last_three_bits_image, cmap='gray')
plt.title('Image using last 3 bits')

plt.subplot(2, 2, (3, 4))
plt.imshow(difference_image, cmap='gray')
plt.title('The Difference Image')

plt.tight_layout()
plt.show()
```
