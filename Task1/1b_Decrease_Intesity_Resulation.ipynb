Importing the libraries

```python
import cv2
import numpy as numpy
import matplotlib.pyplot as plt
```

Load the original image

```python
original_image = cv2.imread('./images/img4.jpg', 0)
gray_image = cv2.resize(original_image, (512, 512))
```

```python
resized_image = gray_image.copy()

img = []

plt.figure(figsize=(20, 12))
plt.subplot(2, 4, 1)
plt.imshow(resized_image, cmap='gray')
plt.title(f"8-bit image")

for i in range(7):
    resized_image = resized_image // 2      # Right-shift each pixel by 1 bit to reduce bit depth
    img.append(resized_image)
    plt.subplot(2, 4, i + 2)
    plt.imshow(img[i], cmap='gray')
    plt.title(f"{8-(i+1)}-bit image")

plt.tight_layout()
plt.show()

    
```
