Importing the libraries

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

Function for Decreasing Spatial Resolution by Half

```python
def resize(image):
    height, width = image.shape
    f = 2   # decreasing factor
    resized_image = np.zeros((height//f, width//f))     # Down-sample the image

    for row in range(0, height, f):
        for col in range(0, width, f):
            resized_image[row//f][col//f] = image[row][col]
            
    return np.uint8(resized_image)
```

Loading the original image

```python
original_image = cv2.imread('./images/img3.webp', 0)
gray_image = cv2.resize(original_image, (512, 512))
```

Displaying the images

```python
plt.figure(figsize=(12, 12))
plt.subplot(3, 3, 1)
plt.imshow(gray_image, cmap='gray')
plt.title(f'Original image {gray_image.shape}')

resized_image = gray_image.copy()
for k in range(8):
    resized_image = resize(resized_image)

    plt.subplot(3, 3, k+2)
    plt.imshow(resized_image, cmap='gray')
    plt.title(f'{resized_image.shape}')

plt.tight_layout()
plt.show()
```
